```python
import wave
import struct
import os
import math
import numpy
from random import randint
pathtolibrary = 'library'
firstarraystore = []
for filename in os.listdir(pathtolibrary):
    opening = wave.open('library/'+ filename, 'rb')
    nframes = opening.getnframes()
    actualframes = opening.readframes(nframes)
    initstore = struct.unpack('%sf'%nframes, actualframes)

    firstarraystore.append(initstore)
secondarraystore = [[]]
keepingtabs = 0
for g in range(len(firstarraystore)):

    for t in range(len(firstarraystore[g])):
        if t % 1000 == 0 and t != 0:
            secondarraystore.append([])
            keepingtabs = keepingtabs + 1
        secondarraystore[keepingtabs].append(firstarraystore[g][t])


songtosum = {}
initsum = []
for i in range(len(secondarraystore)):
    sumly = 0
    for p in secondarraystore[i]:
        if not math.isnan(p):
            sumly = sumly + p
    if numpy.isfinite(sumly):
        av = sumly
        songtosum[av] = secondarraystore[i]
        initsum.append(av)

overallmean = sum(initsum)/len(initsum)
mean = initsum[0]
temporary = [songtosum[initsum[0]]]
ratiodifference = 0
clusters = []
asdofias = []

for t in range(len(initsum)):
    if mean == 0:
        mean = initsum[t+1]

    elif t == len(initsum)-1:
        break
    else:

        ratiodifference = abs((initsum[t]-mean)/mean)




        if ratiodifference < 0.7:
            mean = (mean + initsum[t])/2
            temporary.append(songtosum[initsum[t]])

        else:
            clusters.append(temporary)
            temporary = [[initsum[t]]]
            mean = initsum[t]

new = []
for a in range(len(clusters)):
    new.append([])
    for b in clusters[a]:
        for c in b:
            new[a].append(c)

songtosum1 = {}
initsum1 = []
for i in range(len(new)):
    sumly = 0
    for p in new[i]:
        if not math.isnan(p):
            sumly = sumly + p
    if numpy.isfinite(sumly):
        av = sumly
        songtosum1[av] = new[i]
        initsum1.append(av)



sortedinitsum = sorted(initsum1)
temp = []
temp2 = []
meany = 0
finalsong = [[]]
asala = 0
start = initsum1[randint(0,len(initsum1))]
def hello(start1):
    temp = []
    temp2  = []
    median = 0
    asala = 0
    for e in range(len(sortedinitsum)):
        if sortedinitsum[e] == start1 and e + 0.1*len(initsum1) < len(initsum1) and e - 0.1*len(initsum1)>=0:
            for i in range(1, int(0.1*len(initsum1))):
                temp.append(sortedinitsum[e+i])
                temp.append(sortedinitsum[e -i])
#get the next value for each temp
            for y in temp:
                for r in range(len(initsum1)):
                    if y == initsum1[r] and r < len(initsum1)-1:
                        temp2.append(initsum1[r+1])


            #find mean of temp2
            meany = sum(temp2)/len(temp2)
            initdiff = abs(meany-temp2[0])
            for d in temp2:
                if abs(meany-d) < initdiff:
                    median1 = median
                    median = d
                    initdiff = abs(meany - d)

            finalsong.append(songtosum1[median])
            print(songtosum1[median])
            if len(finalsong)<30:
                hello(median)
            break



hello(start)


#REPEATING PATTERN AVOID:::::::::::::::::::::::::::

finalfinalsong = []
for g in finalsong:
    for t in g:
        finalfinalsong.append(t)

floatlist = [i for i in finalfinalsong]
buf = struct.pack('%sf' % len(finalfinalsong), *floatlist)
p = wave.open('drip.wav', 'wb')
p.setnchannels(2)
p.setframerate(44100)
p.setsampwidth(2)
#p.setsampwidth(len(new_ara))
p.writeframesraw(buf)
p.close()
```
